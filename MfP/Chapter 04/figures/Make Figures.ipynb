```python
def segment(v1,v2):
    x1,y1 = v1
    x2,y2 = v2
    return zip(np.linspace(x1,x2,10),np.linspace(y1,y2,10))

draw(
    
    Points(*segment((1,1),(2,3)),color="C0"),
    Points(*segment((4,2),(2,3)),color="C2"),
    Points(*segment((4,2),(1,1)),color="C3"),
    Points((1,1),color="C0"),
    
    Points(*segment((-3,2),(-1,1)),color="C0"),
    Points(*segment((-3,2),(-2,4)),color="C2"),
    Points(*segment((-1,1),(-2,4)),color="C3"),
    Points((-1,1),color="C0"),
    save_as="4.29.svg"
)
```

```python
draw(
    
    Points(*segment((2,2),(4,1.5)),color="C0"),
    Points(*segment((4,1.5),(1.5,3)),color="C2"),
    Points(*segment((1.5,3),(2,2)),color="C3"),
    Points((2,2),color="C0"),
    
    Points(*[(x*x,y*y) for x,y in segment((2,2),(4,1.5))],color="C0"),
    Points(*[(x*x,y*y) for x,y in segment((4,1.5),(1.5,3))],color="C2"),
    Points(*[(x*x,y*y) for x,y in segment((1.5,3),(2,2))],color="C3"),
    Points((4,4),color="C0"),
    save_as="4.30.svg"
)
```

```python
def arrows4_31():
    for x in range(0,4):
        yield Arrow3D((x+1,0,0),(x,0,0),color='C3')
    yield Segment3D((0,0,0),(4,0,0),color='C3')
    for y in range(0,3):
        yield Arrow3D((4,y+1,0),(4,y,0),color='C2')
    yield Segment3D((4,0,0),(4,3,0),color='C2')    
    for z in range(0,5):
        yield Arrow3D((4,3,z+1),(4,3,z),color='C0')
    yield Segment3D((4,3,0),(4,3,5),color='C0')  
    yield Arrow3D((4,3,5),color='purple')
    
draw3d(
    Box3D(4,3,5),
    *arrows4_31(),
    elev=20,
    save_as='4.31.svg'
)
```
