```python
from math import sqrt, pi

w = (sqrt(2),sqrt(3))
w_scaled = (pi * sqrt(2), pi * sqrt(3))

draw(
    Arrow(w_scaled, color=red),
    Arrow(w, color=black),
    save_as='original_and_scaled_vectors.svg'
)
```

```python
from random import uniform
from vectors import scale
u = (-1,1)
v = (1,1)
def random_r():
    return uniform(-3,3)
def random_s(): 
    return uniform(-1,1)
possibilities = [add(scale(random_r(), u), scale(random_s(), v)) 
                 for i in range(0,500)]
draw(
    Points(*possibilities),
    save_as='lots_of_points.svg'
)
```

```python
def regular_polygon(n):
    return [to_cartesian((1, 2*pi*k/n)) for k in range(0,n)]

draw(
    Polygon(*regular_polygon(7),color='C0'),
    Points(*regular_polygon(7),color='C0'),
    save_as="heptagon.svg"
)
```
